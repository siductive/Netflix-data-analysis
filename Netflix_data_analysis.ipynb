Q. 10) What is the maximum duration of a Movie/Show on Netflix ?

```python
data['duration'].unique()
```

```python
# Convert duration for Movies/TV Shows (Extract numbers and convert to integer)
# Convert duration for Movies (Extract numbers and convert to integer)
data['movie_duration'] = data['duration'].str.extract(r'(\d+)').astype(float)
```

```python
# Separate Movies and TV Shows
movies = data[data['type'] == 'Movie']
tv_shows = data[data['type'] == 'TV Show']
```

```python
# Find the longest Movie
longest_movie = movies.loc[movies['movie_duration'].idxmax()]
```

```python
# Find the longest TV Show
longest_tv_show = tv_shows.loc[tv_shows['movie_duration'].idxmax()]
```

```python
print("Longest Movie:\n", longest_movie[['title', 'duration']])
print("\nLongest TV Show:\n", longest_tv_show[['title', 'duration']])
```
